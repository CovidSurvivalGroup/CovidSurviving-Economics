```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
california = pd.read_csv("../data/new_cali.csv")
california.tail()
```

```python
cali_unemployment = california[["single_home_avg_prices","unemployment"]]
cali_unemployment.plot(kind="line")
plt.title("California")
plt.xlabel("Months 2010-2019")
plt.ylabel("Avg Home Price/unemployment")
```

```python
cali_unemployment = california[["unemployment","employment"]]
cali_unemployment.plot(kind="line")
plt.title("California")
plt.xlabel("Months 2010-2019")
```

```python
cali_unemployment = california[["unemployment rate"]]
cali_unemployment.plot(kind="line")
plt.title("California")
plt.xlabel("Months 2010-2019")
plt.ylabel("unemployment rate")
```

```python
tail=california.tail(5)
post_covid=tail["unemployment"].mean()
```

```python
top=california.head(115)
pre_covid=top["unemployment"].mean()
```

```python
covid_avrg=[pre_covid,post_covid]
x_axis = np.arange(len(covid_avrg))
plt.bar(x_axis, covid_avrg, color='b', alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["pre_covid", "post_covid"])
plt.xlim(-1, len(x_axis)-0)
plt.ylim(0, max(covid_avrg)+50000)
plt.title("Average Unemployment")


```

```python
tail_b=california.tail(5)
post_covid_b=tail_b["single_home_avg_prices"].mean()

top_b=california.head(115)
pre_covid_b=top_b["single_home_avg_prices"].mean()
```

```python
covid_avrg_b=[pre_covid,pre_covid_b]
x_axis_b = np.arange(len(covid_avrg_b))
plt.bar(x_axis_b, covid_avrg_b, color='b', alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Avrg Unemployment", "Avrg Single Home Price"])
plt.xlim(-.50, len(x_axis_b)-.50)
plt.ylim(0, max(covid_avrg_b)+200000)
plt.title("Pre-Covid Era")
```

```python
covid_avrg_c=[post_covid,post_covid_b]
x_axis_c = np.arange(len(covid_avrg_c))
plt.bar(x_axis_c, covid_avrg_c, color='b', alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Avrg Unemployment", "Avrg Single Home Price"])
plt.xlim(-.50, len(x_axis_c)-.50)
plt.ylim(0, max(covid_avrg_c)+50000)
plt.title("Post-Covid Era")
```
