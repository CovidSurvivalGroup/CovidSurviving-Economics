```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
texas = pd.read_csv("new_Texas.csv")
texas.tail()
```

```python
texas_unemployment = texas[["single_home_avg_prices","unemployment"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
plt.ylabel("Avg Home Price/unemployment")                          
```

```python
texas_unemployment = texas[["unemployment","employment"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
```

```python
texas_unemployment = texas[["unemployment rate"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
plt.ylabel("unemployment rate")
```

```python
tail=texas.tail(5)
post_covid=tail["unemployment"].mean()

# top=texas.head(115)
# pre_covid=top["unemployment"].mean()
pre_covid_test=texas.iloc[108:113]
pre_covid=pre_covid_test.mean().unemployment
```

```python
covid_avrg=[pre_covid,post_covid]
x_axis = np.arange(len(covid_avrg))
plt.bar(x_axis, covid_avrg, color='#00A2B4', alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Pre_covid", "covid"])
plt.xlim(-1, len(x_axis)-0)
plt.ylim(0, max(covid_avrg)+50000)
plt.title("Average Unemployment")
```

```python
tail_b=texas.tail(5)
post_covid_b=tail_b["single_home_avg_prices"].mean()

# top_b=texas.head(115)
# pre_covid_b=top_b["single_home_avg_prices"].mean()
pre_covid_test=texas.iloc[108:113]
pre_covid_b=pre_covid_test.mean().single_home_avg_prices
```

```python
covid_avrg_b=[pre_covid,pre_covid_b]
x_axis_b = np.arange(len(covid_avrg_b))
plt.bar(x_axis_b, covid_avrg_b, color=('#00A2B4','r'), alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Avrg Unemployment", "Avrg Single Home Price"])
plt.xlim(-.50, len(x_axis_b)-.50)
plt.ylim(0, max(covid_avrg_b)+50000)
plt.title("Pre-Covid Era")
```

```python
covid_avrg_c=[post_covid,post_covid_b]
x_axis_c = np.arange(len(covid_avrg_c))
plt.bar(x_axis_c, covid_avrg_c, color=('#00A2B4','r'), alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Avrg Unemployment", "Avrg Single Home Price"])
plt.xlim(-.50, len(x_axis_c)-.50)
plt.ylim(0, max(covid_avrg_c)+50000)
plt.title("Covid Era")
```
