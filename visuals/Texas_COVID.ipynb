```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
texas = pd.read_csv("new_Texas.csv")
texas.head()
```

```python
texas_unemployment = texas[["single_home_avg_prices","unemployment"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
plt.ylabel("Avg Home Price/unemployment")                          
```

```python
texas_unemployment = texas[["unemployment","employment"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
```

```python
texas_unemployment = texas[["unemployment rate"]]
texas_unemployment.plot(kind="line")
plt.title("Texas")
plt.xlabel("Months 2010-2019")
plt.ylabel("unemployment rate")
```

```python
tail= texas.tail(5)
post_covid=tail["unemployment"].mean()
post_covid
```

```python
top= texas.head(115)
pre_covid=top["unemployment"].mean()
pre_covid
```

```python
covid_avrg_c= [pre_covid,post_covid]
x_axis_c = np.arange(len(covid_avrg_c))
plt.bar(x_axis_c, covid_avrg_c, color='b', alpha=1, align="center")
tick_locations = [value for value in x_axis_c]
plt.xticks(tick_locations, ['Pre_Covid', 'Post_Covid'])
plt.xlim(-1, len(x_axis_c)-0)
plt.ylim(0, max(covid_avrg_c)+50000)
plt.title('Average Unemployment')
```

```python
covid_avrg= [pre_covid,post_covid]
x_axis = np.arange(len(covid_avrg))
plt.bar(x_axis, covid_avrg, color='b', alpha=1, align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ['Avrg Unemployment', 'Avrg Single Home Price'])
plt.xlim(-.50, len(x_axis)-.50)
plt.ylim(0, max(covid_avrg)+50000)
plt.title('Pre-Covid Era')
```

```python
tail_b= texas.tail(5)
post_covid_b=tail_b["unemployment"].mean()

top_b= texas.head(115)
pre_covid_b=top_b["unemployment"].mean()
```
